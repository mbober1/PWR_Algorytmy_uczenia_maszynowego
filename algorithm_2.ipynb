# Algorytm K najbliższych sąsiadów
Zadaniem poniższego programu jest rozpoznawanie (klasyfikacja) gatunków wina. W tym celu został wykorzystany zbiór danych z bazy scikit-learn, zawierający 178 próbek. Każda z nich składa się z pomiarów 13 cech takich jak zawartość alkoholu czy intensywność koloru. Uczenie maszynowe zostało wykonane z wykorzystaniem algorytmu KNN. Algorytm ten jest jednym z popularniejszych wyborów ze względu na swoją prostą idee oraz intuicyjność. Wykorzystuje on wiedzię o sąsiadujących próbkach, aby decydować o klasyfikacji każdej kolejnej.

# Załadowanie danych
Pierwszym krokiem jest zaimportowanie niezbędnych bilbiotek oraz załadowanie danych.

```python
%config InlineBackend.figure_format = 'svg'

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from matplotlib import pyplot as plt
from joblib import dump, load
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np

# load data
x, y = load_wine(return_X_y=True, as_frame=True)
```

# Opis danych
Z użyciem prostej metody "describe" możemy otrzymać zestawienie najważniejszych parametrów dotyczących użytego zbioru danych.

# Normalizacja i podział danych, uczenie, raport, confusion_matrix
Kolejnym krokiem powinna być normalizacja danych. Dzięki niej wszystkie wartości będą mieścić się w przedziale [0 1]. 

Następnie został przeprowadzony podział zbioru danych na dane treningowe oraz dane testowe. Jest to niezwykle ważne ponieważ bez tego kroku niemożiwa byłaby rzetelna analiza skuteczności algorytmu sztucznej inteligencji. 

Dalej dane zostały poddane sortowaniu. Nie jest to krok niezbędny, ale poprawia on estetykę niżej prezentowanych wykresów.

Najważnejszym krokiem jest stworzenie obiektu SVC oraz wykonanie na nim metody "fit" przeprowadzającej docelowe uczenie. Zgodnie z wytycznymi projektu, program zawiera tryb treningowy oraz tryb normalny. Tryb treningowy zostaje uruchomiony, gdy w folderze nie znajduje się plik "model_2.joblib". W przeciwnim wypadku uruchamia się trym normalny, który ładuje gotowy model z pliku.

Po otrzymaniu modelu, program przystępuje do predykcji z wykorzystaniem danych testowych. Otrzymane wyniki w formie "confusion matrix" są przedstawione poniżej (wersja tekstowa oraz graficzna).

```python
# normalize values to same range (0, 1)
scaler = MinMaxScaler() 
scaled = scaler.fit_transform(x)
x = pd.DataFrame(scaled)

# split data to train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=42)

# sort values
x_test['y'] = y_test
x_test = x_test.sort_values(by=['y'])
y_test = x_test['y']
x_test = x_test.drop(columns=['y'])

# create classifier and learn it
try:
  clf = load('model_2.joblib')
except:
  clf = KNeighborsClassifier()
  clf.fit(x_train, y_train)
  dump(clf, 'model_2.joblib')

# predict
y_predicted = clf.predict(x_test)

print(classification_report(y_test, y_predicted))
print(confusion_matrix(y_test, y_predicted))
```

### Ładniejszy confusion_matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test, cmap=plt.cm.Blues)
plt.show()
```

# Wykres jakości klasyfikacji
Poniżej znajduje się wykres przedstawiający wyniki predykcji algorytmu na danych testowych. Na osi poziomej znajdują się próbki, a na osi pionowej klasa wina do której została zaklasyfikowana dana próbka. Większość próbek została zakwalifikowana poprawnie, co potwierdza "confusion matrix". # Zależność od ilości sąsiadów
Na poniższym wykresie zostały przedstaione wyniki badania skuteczności algorytmu ze względu na wybraną ilość sąsiadów branych pod uwagę podczas klasyfikacji. Można zauważyć że najlepsze wyniki otrzymano dla pięciu sąsiadów. Nie bez przyczyny jest to wartość domyślna dla tego parametru.

```python
neighbors = range(1, 11)
scores = []

for neighbor in neighbors:
  new_clf = KNeighborsClassifier(n_neighbors=neighbor)
  new_clf.fit(x_train, y_train)
  y_predicted = new_clf.predict(x_test)
  score = accuracy_score(y_test, y_predicted)
  scores.append(score)

fig = plt.figure()
ax = fig.add_subplot()
ax.grid()
ax.plot(neighbors, scores, "o-")
plt.title("Dependency on neighbors")
plt.xlabel("Neighbor values")
plt.ylabel("Accuracy score")
```

# Wizualizacja zbioru danych w zależności od funkcji wagi
Przy wykorzystaniu tego algorytmu mamy do dyspozycji dwie fukncje wagi. 

- Jednolitą, gdzie wszystkie punkty w każdym sąsiedztwie są jednakowo ważne.
- Odległościową, w której wraz ze wzrostem odległości sąsiedztwa maleje jego znaczenie dla końcowego wyniku.

Wyniki badania obu powyższych zależności przedstawiono na wykresach.

Domyślnym wyborem jest funkcja jednolita, jednakże z poniższych wykresów można dojść do wniosku że fukncja odległościowa jest dokładniejsza, ponieważ lepiej klasyfikuje tak zwanych "outsiderów". Z drugiej jednak strony jest ona bardziej podatna na błedy danych wejściowych.

```python
def make_meshgrid(x, y, step = 0.02):
  x_min, x_max = x.min() - 0.1, x.max() + 0.1
  y_min, y_max = y.min() - 0.1, y.max() + 0.1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
  return xx, yy

def plot_contours(axis, clf, x, y, **params):
  Z = clf.predict(np.c_[x.ravel(), y.ravel()])
  Z = Z.reshape(x.shape)
  out = axis.contourf(x, y, Z, **params)
  return out


x_train_np = x_train.to_numpy()[:, :2]
y_train_np = y_train.to_numpy()

models = (
  KNeighborsClassifier(weights="uniform"),
  KNeighborsClassifier(weights="distance"),
)

models = (clf.fit(x_train_np, y_train_np) for clf in models)

titles = (
  "KNeighborsClassifier with uniform weights",
  "KNeighborsClassifier with distance weights",
)

X0 = x_train_np[:, 0]
X1 = x_train_np[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title in zip(models, titles):
  fig, ax = plt.subplots()
  plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
  ax.scatter(X0, X1, c=y_train_np, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
  plt.title(title)
  plt.xlabel("Alcohol value")
  plt.ylabel("Malic acid value")

plt.show()
```

Poniżej prezentowane są wyniki wykorzystania tych algorytmów. W prezentowanych przypadku wszystkie wyniki są do siebie mocno zbliżone.

```python
models = (
  KNeighborsClassifier(algorithm="ball_tree"),
  KNeighborsClassifier(algorithm="kd_tree"),
  KNeighborsClassifier(algorithm="brute"),
)

x_train_np = x_train.to_numpy()[:,  7:9]
y_train_np = y_train.to_numpy()
models = (clf.fit(x_train_np, y_train_np) for clf in models)

titles = (
  "KNeighborsClassifier algorithm BallTree",
  "KNeighborsClassifier algorithm KDTree",
  "KNeighborsClassifier algorithm Brute Force",
)


X0 = x_train_np[:, 0]
X1 = x_train_np[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title in zip(models, titles):
  fig, ax = plt.subplots()
  plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
  ax.scatter(X0, X1, c=y_train_np, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
  plt.title(title)
  plt.xlabel("Alcohol value")
  plt.ylabel("Malic acid value")

plt.show()
```

# Krzywa uczenia
Ostatnim wykresem dla tego algorytmu jest karzywa uczenia, dzięki której jesteśmy w stanie ocenić czy nasz model nie jest przeuczony.

```python

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import learning_curve


def plot_learning_curve(
    estimator,
    title,
    X,
    y,
    axes=None,
    ylim=None,
    cv=None,
    n_jobs=None,
    train_sizes=np.linspace(0.1, 1.0, 5),
):
    if axes is None:
        _, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].set_title(title)
    if ylim is not None:
        axes[0].set_ylim(*ylim)
    axes[0].set_xlabel("Training examples")
    axes[0].set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(
        estimator,
        X,
        y,
        cv=cv,
        n_jobs=n_jobs,
        train_sizes=train_sizes,
        return_times=True,
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Plot learning curve
    axes[0].grid()
    axes[0].fill_between(
        train_sizes,
        train_scores_mean - train_scores_std,
        train_scores_mean + train_scores_std,
        alpha=0.1,
        color="r",
    )
    axes[0].fill_between(
        train_sizes,
        test_scores_mean - test_scores_std,
        test_scores_mean + test_scores_std,
        alpha=0.1,
        color="g",
    )
    axes[0].plot(
        train_sizes, train_scores_mean, "o-", color="r", label="Training score"
    )
    axes[0].plot(
        train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score"
    )
    axes[0].legend(loc="best")

    # Plot n_samples vs fit_times
    axes[1].grid()
    axes[1].plot(train_sizes, fit_times_mean, "o-")
    axes[1].fill_between(
        train_sizes,
        fit_times_mean - fit_times_std,
        fit_times_mean + fit_times_std,
        alpha=0.1,
    )
    axes[1].set_xlabel("Training examples")
    axes[1].set_ylabel("fit_times")
    axes[1].set_title("Scalability of the model")

    # Plot fit_time vs score
    fit_time_argsort = fit_times_mean.argsort()
    fit_time_sorted = fit_times_mean[fit_time_argsort]
    test_scores_mean_sorted = test_scores_mean[fit_time_argsort]
    test_scores_std_sorted = test_scores_std[fit_time_argsort]
    axes[2].grid()
    axes[2].plot(fit_time_sorted, test_scores_mean_sorted, "o-")
    axes[2].fill_between(
        fit_time_sorted,
        test_scores_mean_sorted - test_scores_std_sorted,
        test_scores_mean_sorted + test_scores_std_sorted,
        alpha=0.1,
    )
    axes[2].set_xlabel("fit_times")
    axes[2].set_ylabel("Score")
    axes[2].set_title("Performance of the model")

    return plt

cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)


plot_learning_curve(clf, 'Learning curves', x, y, cv=cv, n_jobs=4)
```
